Представим матрицу $A \in R^{nxn}$ в блочном виде

\begin{equation}
A = 
\begin{bmatrix} 
A_{11} & A_{12} \\
A_{21} & A_{22}\\
\end{bmatrix}
\end{equation}
$A_{11}$ - подматрица матрицы A  - размера rxr, $A_{12}$ - размера rx(n-r), $A_{21}$ - размера (n-r) x r, $A_{22}$- размера  (n-r) x (n-r). Компоненты L и U искомого разложения также запишем в блочном виде

\begin{equation}
L = 
\begin{bmatrix} 
L_{11} & 0 \\
L_{21} & L_{22}\\
\end{bmatrix}
\quad
U = 
\begin{bmatrix} 
U_{11} & U_{12} \\
0 & U_{22}\\
\end{bmatrix}
\quad
\end{equation}

где $L_{11}, L_{21}, L_{22}, U_{11}, U_{12}, U_{22}$ - соответствующего разиера подматрицы матриц L и U.


Имеем
\begin{equation}
\begin{bmatrix} 
A_{11} & A_{12} \\

# Problem 3

Решить задачу наименьших квадратов

- Реализовать QR-разложение используя Householder отражения. Показать корректность работы на примере относительных ошибок определения матриц $Q, R$ и $A = QR$
- Сгенерировать матрицу $A$ и вектор $b$, решить переопределенную систему уравнений используя QR-разложение и решение уравнения, полученного из условия оптимальности первого порядка.
- Привести график зависимости точности решения от числа обусловленности матрицы системы для случаев из пункта (2).

```python
import numpy as np
import scipy.linalg as linalg 
def column_convertor(x):
    x.shape = (1, x.shape[0])
    return x

def get_norm(x):
    return np.sqrt(np.sum(np.square(x)))

def householder_transformation(v):
    size_of_v = v.shape[1]
    e1 = np.zeros_like(v)
    e1[0, 0] = 1
    vector = get_norm(v) * e1
    if v[0, 0] < 0:
        vector = - vector
    u = (v + vector).astype(np.float32)
    H = np.identity(size_of_v) - ((2 * np.matmul(np.transpose(u), u)) / np.matmul(u, np.transpose(u)))
    return H

def qr_step_factorization(q, r, iter, n):
    v = column_convertor(r[iter:, iter])
    Hbar = householder_transformation(v)
    H = np.identity(n)
    H[iter:, iter:] = Hbar
    r = np.matmul(H, r)
    q = np.matmul(q, H)
    return q, r

def main():
    n, m = list(map(int, input("Write size of the matrix in space separated format\n").split()))
    A = np.random.rand(n,m)
    print('The random matrix is \n', A)
    Q = np.identity(n)
    R = A.astype(np.float32)
    for i in range(min(n, m)):
        Q, R = qr_step_factorization(Q, R, i, n)
    min_dim = min(m, n)
    R = np.around(R, decimals=6)
    R = R[:min_dim, :min_dim]
    Q = np.around(Q, decimals=6)
    print('A after QR factorization')
    print('R matrix')
    print(R, '\n')
    print('Q matrix')
    print(Q)
if __name__ == "__main__":
    main()
```
